## Task 1

```r
id = c(0, 2, 1)
repeats = 1000
samples_sizes = c(100, 200, 1000)

find_pattern <- function (sample, id) {
    sample_length <- length(sample)
    counter = 0
    for (i in 1:(sample_length - 2)) {
        if (id[1] == sample[i] & id[2] == sample[i + 1] & id[3] == sample[i + 2]) {
            counter = counter + 1
        }
    }
    return (counter)
}



for (size in samples_sizes) {
    results <- replicate(
        repeats,
        result <- find_pattern(sample(0:9, size, replace=T), id)
    )

    success <- length(which(results > 0))
    cat("Successes for n = ")
    cat(size)
    cat(" === ")
    cat(success)
    cat(". LLN ")
    cat(success / repeats)
    cat("\n")
}
```

## Task 3

```r
id = c(0, 2, 1)
big_seq_size = 25000
repeats = 1000

find_pattern <- function (sample, id) {
    sample_length <- length(sample)
    counter = 0
    for (i in 1:(sample_length - 2)) {
        if (id[1] == sample[i] & id[2] == sample[i + 1] & id[3] == sample[i + 2]) {
            counter = counter + 1
        }
    }
    return (counter)
}

calc <- function () {
    seq = sample(0:9, big_seq_size, replace=T)

    for (i in 1:(length(seq) - 2)) {
        if (id[1] == seq[i] & id[2] == seq[i + 1] & id[3] == seq[i + 2]) {
            return (i);
        }
    }
}

results <- replicate(
    repeats,
    result <- calc()
)

cat("Mean: ")
cat(mean(results))
```

## Task 4

```r
N = 1000
n_values = c(5, 10, 20)

for (n in n_values) {
    a <- colMeans(matrix(sample(1:3, N*n, replace = TRUE),N))
    cat(a)
    cat("\n")
    
    Fd=ecdf(a)
    
    values = seq(min(a), max(a), length=100)
    plot(x=values, y=Fd(values))
    lines(values, pnorm(values, mean=mean(a), sd=sd(a)))

    sigma_values = seq(mean(a) - 3*sd(a), mean(a) + 3*sd(a), length=200)
    b <- max(abs(Fd(sigma_values) - pnorm(sigma_values, mean=mean(a), sd=sd(a))))
    cat(b)
    cat("\n")
}
```
